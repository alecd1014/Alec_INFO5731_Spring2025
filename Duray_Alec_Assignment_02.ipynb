# Question 1 (25 points)

Write a python program to collect text data from **either of the following sources** and save the data into a **csv file:**

(1) Collect all the customer reviews of a product (you can choose any porduct) on amazon. [atleast 1000 reviews]

(2) Collect the top 1000 User Reviews of a movie recently in 2023 or 2024 (you can choose any movie) from IMDB. [If one movie doesn't have sufficient reviews, collect reviews of atleast 2 or 3 movies]


(3) Collect the **abstracts** of the top 10000 research papers by using the query "machine learning", "data science", "artifical intelligence", or "information extraction" from Semantic Scholar.

(4) Collect all the information of the 904 narrators in the Densho Digital Repository.

(5)**Collect a total of 10000 reviews** of the top 100 most popular software from G2 and Capterra.


```python
# Your code here
from bs4 import BeautifulSoup
from urllib.request import Request, urlopen
import pandas as pd
import re

#create list to add each section of information
narrator_data = []
#create loop for each page of narrators
for i in range(1,41):
  url = f'https://ddr.densho.org/narrators/?page={i}'
  #get the content of the page from the url
  header = {'User-Agent': 'Mozilla/5.0'}
  req = Request(url, headers=header)
  page = urlopen(req).read()
  #use beautiful soup to scrape through the content on the page
  soup = BeautifulSoup(page, 'html.parser')
  #find all the individual sections of each narrator
  table = soup.find_all('div', class_='col-md-6 col-sm-6')

  #iterate through each row to find the text for each narrator and add it to the created list
  for row in table:
    info = row.find('div', attrs = {'class':'source muted'})
    narrator_data.append(info.text)

#remove blank space from each item in the list
narrator_data = [' '.join(string.split()) for string in narrator_data]
#create a dataframe from the list of reviews
narrator_df = pd.DataFrame({'Narrator Data':narrator_data})
#name the csv file
csv_file = 'narrators.csv'
#convert the dataframe to a csv file
narrator_df.to_csv(csv_file, index=False)
```

# Question 2 (15 points)

Write a python program to **clean the text data** you collected in the previous question and save the clean data in a new column in the csv file. The data cleaning steps include: [Code and output is required for each part]

(1) Remove noise, such as special characters and punctuations.

(2) Remove numbers.

(3) Remove stopwords by using the stopwords list.

(4) Lowercase all texts

(5) Stemming.

(6) Lemmatization.

```python
# Write code for each of the sub parts with proper comments.
#Part 1
import pandas as pd
#convert the csv file into a dataframe
all_narrators = pd.read_csv('narrators.csv')

#remove all special characters and punctuation from each narrator section
#also add a new column to the dataframe for the clean data
all_narrators['Clean data'] = all_narrators['Narrator Data'].str.replace('[^\w\s]', '', regex=True)
#print head of dataframe
all_narrators.head()
```

```python
#Part 2

#remove all numbers from each narrator section
all_narrators['Clean data'] = all_narrators['Clean data'].str.replace('\d+', '', regex=True)
#print head of dataframe
all_narrators.head()
```

```python
#Part 3

import nltk
#download list of stopwords
nltk.download('stopwords')
from nltk.corpus import stopwords
#create list of english stopwords
stop = stopwords.words('english')

#remove stopwords from each narrator section
all_narrators['Clean data'] = all_narrators['Clean data'].fillna('').apply(lambda x: " ".join(x for x in x.split() if x not in stop))
#print head of dataframe
all_narrators.head()
```

```python
#Part 4

#convert all text in each narrator section to lowercase
all_narrators['Clean data'] = all_narrators['Clean data'].apply(lambda x: " ".join(x.lower() for x in x.split()))
#print head of dataframe
all_narrators.head()
```

```python
#Part 5

from nltk.stem import PorterStemmer
#import stemmer module
st = PorterStemmer()

#use stemming on each narrator section
all_narrators['Clean data'] = all_narrators['Clean data'].apply(lambda x: " ".join([st.stem(word) for word in x.split()]))
#print head of dataframe
all_narrators.head()
```

```python
#Part 6

from textblob import Word
import nltk
#download module for lemmatization
nltk.download('wordnet')

#use lemmatization on each narrator section
all_narrators['Clean data'] = all_narrators['Clean data'].apply(lambda x: " ".join([Word(word).lemmatize() for word in x.split()]))
#print head of dataframe
print(all_narrators.head())
#add the clean data to the csv file as a new column
all_narrators.to_csv(csv_file, index=False)
```

```python
# Your code here

#Part 1
import nltk
from nltk.tokenize import word_tokenize
from nltk.tag import pos_tag
from collections import Counter

nltk.download('averaged_perceptron_tagger_eng')
#tag each word in the data and attach its relevant part of speech
pos_tags = pos_tag(all_narrators['Clean data'])
#count the number of each different part of speech
count = Counter(j for i, j in pos_tags)
#calculate the number of nouns
n = count['NN'] + count['NNS'] + count['NNP']
#calculate the number of verbs
v = count['VBP'] + count['VBD'] + count['VBZ']
#calculate the number of adjectives
adj = count['JJ'] + count['JJR'] + count['JJS']
#calculate the number adverbs
adv = count['RB']
#print the results
print(f'Number of nouns: {n}')
print(f'Number of verbs: {v}')
print(f'Number of adjectives: {adj}')
print(f'Number of adverbs: {adv}')
```

```python
#Part 2
import nltk
from nltk import CFG
from nltk.tokenize import word_tokenize
from nltk import pos_tag

nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')
#form grammar to define sentence parts
gram = CFG.fromstring("""
  S -> NP VP
  NP -> Det N | Det Adj N | N
  VP -> V NP | V NP PP
  PP -> P NP
""")
parser = nltk.ChartParser(gram)

def sentence_pos_tags(sentence):
  tokens = word_tokenize(sentence)
  pos_tags = pos_tag(tokens)
  #convert sentence parts
  pos_tag_map = {'DT': 'Det', 'JJ': 'Adj', 'NN': 'N', 'VB': 'V', 'IN': 'P'}
  simp_tags = [(word, pos_tag_map.get(tag, tag)) for word, tag in pos_tags]
  new_tokens = [tag for word, tag in simp_tags]
  return new_tokens

sentence_list = all_narrators['Clean data'].values.tolist()

con_parse = []
for i in range(len(sentence_list)):
  sentence = sentence_list[i]#.split()
  new_sentence = sentence_pos_tags(sentence)
  con_parse.append(new_sentence)

print(con_parse[0])
```

Github MarketPlace page:

```python
from bs4 import BeautifulSoup
from urllib.request import Request, urlopen
import pandas as pd
import time
#Part 1
#create lists for each section of data
product_name = []
description = []
product_url = []
page_number = []
#since the first page does not have a page designation in the url, the items on this page need to be
#done separately

#define the target url
git_url = 'https://github.com/marketplace?type=actions'
#request the url using the user agent
req = Request(git_url, headers={'User-Agent': 'Mozilla/5.0'})
#open and read the read the url
page = urlopen(req).read()
#use beautiful soup and html parser to parse the page
soup = BeautifulSoup(page, 'html.parser')

#find all the instances of a product on the page
table = soup.find_all('div', class_='position-relative border rounded-2 d-flex marketplace-common-'\
                      'module__marketplace-item--MohVH gap-3 p-3')
#iterate through each product to find the name of the product, the product description, and the product url
for row in table:
  name = row.find('a', attrs = {'class':'marketplace-common-module__marketplace-item-link--jrIHf l'\
                    'ine-clamp-1'}).get_text(strip=True) #strip out unneeded whitespace
  desc = row.find('p', attrs = {'class': 'mt-1 mb-0 text-small fgColor-muted line-clam'\
                                  'p-2'}).get_text(strip=True) #strip out unneeded whitespace
  prod_url = 'https://github.com' + row.find('a')['href'] #add the beginning of the url to what is given in html code
  product_name.append(name) #add name to name list
  description.append(desc) #add description to description list
  product_url.append(prod_url) #add url to url list
  page_number.append(1) #since this is only the first page add 1 for page number
#for rest of pages the url is the same except for the page number
#so we can use a loop

#start the page number counter
pn = 2
#run the loop through page 50 to get 1000 products since there are 20 products on each page
for i in range(2, 51):
  #define the new url leaving the page number blank
  new_url = f'https://github.com/marketplace?page={i}&type=actions'
  #the rest is the same as above
  req = Request(new_url, headers={'User-Agent': 'Mozilla/5.0'})
  page = urlopen(req).read()
  soup = BeautifulSoup(page, 'html.parser')
  table = soup.find_all('div', class_='position-relative border rounded-2 d-flex marketplace-common-'\
                      'module__marketplace-item--MohVH gap-3 p-3')
  for row in table:
    name = row.find('a', attrs = {'class':'marketplace-common-module__marketplace-item-link--jrIHf l'\
                    'ine-clamp-1'}).get_text(strip=True)
    desc = row.find('p', attrs = {'class': 'mt-1 mb-0 text-small fgColor-muted line-clam'\
                                  'p-2'}).get_text(strip=True)
    prod_url = 'https://github.com' + row.find('a')['href']
    product_name.append(name)
    description.append(desc)
    product_url.append(prod_url)
    page_number.append(pn)
  #add 1 to the page number before going on to the next page
  pn = pn + 1
  #add a pause to prevent server overload
  time.sleep(2)

#create a dataframe using the lists of the products
gitdata_df = pd.DataFrame({'Product Name': product_name, 'Description': description, 'URL': product_url
                           , 'Page number': page_number})
#create the name of the csv file
csv_data = 'gitdata.csv'
#create the csv using the dataframe
gitdata_df.to_csv(csv_data, index=False)
```

```python
#Part 2
import re
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
nltk.download('wordnet')
from textblob import Word
#convert csv file into dataframe
all_gitdata = pd.read_csv('gitdata.csv')
#the url column and the page number column do not need any preprocessing changes
#the product description can use them however

#remove the numbers from the product description column
all_gitdata['Description'] = all_gitdata['Description'].str.replace('[^\w\s]', '', regex=True)

#remove stopwords from the product description column
stop = stopwords.words('english')
all_gitdata['Description'] = all_gitdata['Description'].fillna('').apply(
                            lambda x: " ".join(x for x in x.split() if x not in stop))

#convert all text to lowercase
all_gitdata['Description'] = all_gitdata['Description'].apply(
                            lambda x: ' '.join(x.lower() for x in x.split()))

#use lemmatization on product description column
all_gitdata['Description'] = all_gitdata['Description'].apply(
                            lambda x: ' '.join([Word(word).lemmatize() for word in x.split()]))

#replace the product description column with the new product description column with the clean data
all_gitdata.to_csv(csv_data, index = False)
```
